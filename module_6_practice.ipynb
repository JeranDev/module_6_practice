 Question 12

Fix the error shown above using the method of your preference. You may need to use ChatGPT and/or Google to figure out how to do this (hint: use the .str.title() method).

How many computers in this data set are from the company Apple?


```python
df['Company'] = df['Company'].str.lower()

# Standardize company names
company_mapping = {
    'hp': 'HP',
    'dell': 'Dell',
    'lenovo': 'Lenovo',
    'asus': 'Asus',
    'acer': 'Acer',
    'toshiba': 'Toshiba',
    'msi': 'MSI',
    'apple': 'Apple',
    'samsung': 'Samsung',
    'xiaomi': 'Xiaomi',
    'vero': 'Vero',
    'razer': 'Razer',
    'mediacom': 'Mediacom',
    'google': 'Google',
    'fujitsu': 'Fujitsu',
    'lg': 'LG',
    'microsoft': 'Microsoft',
    'huawei': 'Huawei',
    'chuwi': 'Chuwi'
}

df['Company'] = df['Company'].map(company_mapping)

# Check the standardized counts
print(df['Company'].value_counts())
```

 Question 13

It's difficult to tell, but some of the laptops seem to have extremely high prices. Use the z-score method with a threshold of 3 to create a filter and determine which laptops are outliers in the data set in terms of price.

Using the outlier detection method from above, how many outliers exist in this data set according to price?


```python
average_price = df['Price'].mean()
std_dev_price = df['Price'].std()

df['Z-Score'] = (df['Price'] - average_price) / std_dev_price
```

```python
df[df['Z-Score'] > 3]
```
